```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

file_path = '/content/drive/MyDrive/Colab Notebooks/LY DV Practicles/Car_Info.csv'

# Sample data
data = pd.read_csv(file_path)

# Example: Create a histogram using Matplotlib
plt.hist(data['highway'], bins=10, color='blue', alpha=0.7)
plt.xlabel('Numeric Column')
plt.ylabel('Frequency')
plt.title('Histogram of Numeric Data')
plt.show()


# Example: Create a kernel density plot using Seaborn
sns.kdeplot(data['Popularity'], shade=True, color='green')
plt.xlabel('Numeric Column')
plt.ylabel('Density')
plt.title('Kernel Density Plot of Numeric Data')
plt.show()
```
